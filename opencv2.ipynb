```python
import cv2
import numpy as np
import os


# Function to extract shape features from an image
def extract_shape_features(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour = max(contours, key=cv2.contourArea)
    perimeter = cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, 0.01 * perimeter, True)
    return np.array([len(approx)])

# Function to extract color features from an image
def extract_color_features(image):
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    hist = cv2.normalize(hist, hist).flatten()
    return hist.reshape(-1, 1)

# Function to extract shape and color features from an image
def retrieve_images(query_image_path, dataset_path):
    # Load query image and extract features
    query_image = cv2.imread(query_image_path)
    query_shape_features = extract_shape_features(query_image).reshape(-1, 1)
    query_color_features = extract_color_features(query_image)

    # Load dataset images and extract features
    dataset_images = []
    dataset_shape_features = []
    dataset_color_features = []
    for image_path in os.listdir(dataset_path):
            image = cv2.imread(os.path.join(dataset_path, image_path))
            shape_features = extract_shape_features(image).reshape(-1, 1)
            color_features = extract_color_features(image)
            dataset_images.append(image)
            dataset_shape_features.append(shape_features)
            dataset_color_features.append(color_features)

    # Convert features to numpy arrays
    dataset_shape_features = np.array(dataset_shape_features)
    dataset_color_features = np.array(dataset_color_features)

    # Compute similarity scores for shape and color features separately
    shape_scores = np.abs(dataset_shape_features - query_shape_features)

from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity instead of Euclidean distance
    distances = 1 - cosine_similarity(query_features.reshape(1, -1), dataset_features)

    # Combine shape and color scores using weighted sum
    alpha = 0.5
    beta = 1 - alpha
    scores = alpha * shape_scores + beta * color_scores

    # Get indices of top-k similar images
    k = 5
    print(scores.shape)
    indices = np.argsort(scores).astype(int)
    #print(indices)
    indices = indices.ravel()
    if len(indices) > 0:
    #Load top-k similar images and display them
        for i in indices:
            print(i)
            image = dataset_images[int(indices[i])]
            cv2.imshow(f"Similar image {i+1}", image)

        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
       print("No similar images found.")
        
query_path = r"D://Downloads//i10new.jpg"
database_path = r"D:\FYP\True\IVDB\cars"
retrieve_images(query_path, database_path)
```

```python
import cv2
import numpy as np
from skimage.feature import greycomatrix, greycoprops
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity

class HuMoments:
    def __init__(self, num_moments=7):
        self.num_moments = num_moments

    def describe(self, image):
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        moments = cv2.moments(image).values()
        hu_moments = cv2.HuMoments(moments).flatten()
        return hu_moments[:self.num_moments]

class ColorDescriptor:
    def __init__(self, bins=(8, 8, 8)):
        self.bins = bins

    def describe(self, image):
        image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([image], [0, 1, 2], None, self.bins, [0, 180, 0, 256, 0, 256])
        hist = cv2.normalize(hist, hist).flatten()
        return hist

class CBIR:
    def __init__(self, shape_features, color_features):
        self.shape_features = shape_features
        self.color_features = color_features

    def search(self, query_path, dataset, k=5):
        query_image = cv2.imread(query_path)
        query_shape_features = self.shape_features.describe(query_image)
        query_color_features = self.color_features.describe(query_image)
        query_features = np.concatenate((query_shape_features, query_color_features))

        dataset_shape_features = np.asarray([self.shape_features.describe(image) for image in dataset])
        dataset_color_features = np.asarray([self.color_features.describe(image) for image in dataset])
        dataset_features = np.concatenate((dataset_shape_features, dataset_color_features), axis=1)

        distances = 1 - cosine_similarity(query_features.reshape(1, -1), dataset_features)[0]
        indices = np.argsort(distances)[:k]
        print(indices)
     
```
